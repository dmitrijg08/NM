```python
import numpy as np

# Таблиця значень функції
x_values = np.array([2.4, 2.6, 2.8, 3, 3.2, 3.4, 3.6, 3.8, 4, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Крок таблиці
h = x_values[1] - x_values[0] # h = 0.2

# Обчислення кінцевих різниць
def finite_differences(y):
    n = len(y)
    delta_y = np.zeros((n, n))
    delta_y[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            delta_y[i, j] = delta_y[i + 1, j - 1] - delta_y[i, j - 1]

    return delta_y

# Отримуємо таблицю кінцевих різниць
delta_table = finite_differences(y_values)

# Виводимо таблицю кінцевих різниць для наочності
print("Таблиця кінцевих різниць:")
np.set_printoptions(precision=4, suppress=True)
print(delta_table)

d1 = delta_table[2, 1]   # Δy
d2 = delta_table[2, 2]   # Δ²y
d3 = delta_table[2, 3]   # Δ³y
d4 = delta_table[2, 4]   # Δ⁴y
d5 = delta_table[2, 5]   # Δ⁵y
d6 = delta_table[2, 6]   # Δ⁶y
d7 = delta_table[2, 7]   # Δ⁷y
d8 = delta_table[2, 8]   # Δ⁸y
d9 = delta_table[2, 9]   # Δ⁹y

# Функція для обчислення першої похідної
def first_derivative(d1, d2, d3, d4, d5, d6, d7, d8, d9, h):
    terms = (
        d1
        - d2 / 2
        + d3 / 3
        - d4 / 4
        + d5 / 5
        - d6 / 6
        + d7 / 7
        - d8 / 8
        + d9 / 9
    )
    return terms / h

# Функція для обчислення другої похідної (залишаємо як у зразку, до 4-ї різниці)
def second_derivative(d2, d3, d4, h):
    # Формула y'' ≈ (1/h²) * (Δ²y - Δ³y + (11/12)Δ⁴y)
    return (d2 - d3 + 11 * d4 / 12) / (h**2)

# Обчислюємо першу та другу похідні в точці x = 2.8 (ЗМІНЕНО)
y1_prime = first_derivative(d1, d2, d3, d4, d5, d6, d7, d8, d9, h)
y1_double_prime = second_derivative(d2, d3, d4, h) # Використовуємо d2, d3, d4

# Округляємо до точності 0.001 (НЕ ЗМІНЮЄМО)
y1_prime_rounded = round(y1_prime, 3)
y1_double_prime_rounded = round(y1_double_prime, 3)

# Виводимо результати (ЗМІНЕНО, виправляємо точку x)
print(f"\nТочка обчислення x = {x_values[2]}")
print(f"Перша похідна y'({x_values[2]}) ≈ {y1_prime_rounded} (точне: {y1_prime})")
print(f"Друга похідна y''({x_values[2]}) ≈ {y1_double_prime_rounded} (точне: {y1_double_prime})")

# ----------------- КІНЕЦЬ ЗМІН -----------------
```

```python
import numpy as np

# Таблиця значень функції (РОЗКОМЕНТОВАНО)
x_values = np.array([2.4, 2.6, 2.8, 3, 3.2, 3.4, 3.6, 3.8, 4, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Крок таблиці
h = x_values[1] - x_values[0]

# Перша похідна
first_derivative = (y_values[1:] - y_values[:-1]) / h

# Друга похідна
second_derivative = (first_derivative[1:] - first_derivative[:-1]) / h

# x = 2.8 є 3-м елементом, тому має індекс 2
target_index = 2
target_x_value = x_values[target_index]

# Виведення результатів
print("Перша похідна:")
print(f"x = {target_x_value}:", round(first_derivative[target_index],3))  # Змінено індекс з 0 на 2

print("Друга похідна:")
print(f"x = {target_x_value}:", round(second_derivative[target_index], 3))  # Змінено індекс з 0 на 2
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Таблиця значень функції
x_values = np.array([2.4, 2.6, 2.8, 3, 3.2, 3.4, 3.6, 3.8, 4, 4.2, 4.4, 4.6])
y_values = np.array([3.526, 3.782, 3.945, 4.043, 4.104, 4.155, 4.222, 4.331, 4.507, 4.775, 5.159, 5.683])

# Крок таблиці
h = x_values[1] - x_values[0]

# Масив first_derivative буде мати на 1 елемент менше (N-1)
first_derivative = (y_values[1:] - y_values[:-1]) / h

# Масив second_derivative буде мати на 2 елементи менше (N-2)
second_derivative = (first_derivative[1:] - first_derivative[:-1]) / h

# Створюємо фігуру для графіків
plt.figure(figsize=(12, 10))

# --- 1. Графік функції y(x) ---
plt.subplot(3, 1, 1)
# Інтерполяція для гладкої лінії
x_fine_y = np.linspace(x_values[0], x_values[-1], 500)
y_fine = np.interp(x_fine_y, x_values, y_values)

plt.plot(x_fine_y, y_fine, label='Функція y(x)', color='blue')
plt.scatter(x_values, y_values, color='red', label='Вхідні дані')
plt.title('Функція y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
# Важливо: first_derivative (N-1 елементів) відповідає x_values[:-1] (від 2.4 до 4.4)
x_data_d1 = x_values[:-1]
x_fine_d1 = np.linspace(x_data_d1[0], x_data_d1[-1], 500)
first_deriv_fine = np.interp(x_fine_d1, x_data_d1, first_derivative)

plt.plot(x_fine_d1, first_deriv_fine, label="Перша похідна y'(x) (2. спосіб)", color='orange')
plt.title("Перша похідна y'(x)")
plt.xlabel('x')
plt.ylabel("y'")
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
# Важливо: second_derivative (N-2 елементів) відповідає x_values[:-2] (від 2.4 до 4.2)
x_data_d2 = x_values[:-2]
x_fine_d2 = np.linspace(x_data_d2[0], x_data_d2[-1], 500)
second_deriv_fine = np.interp(x_fine_d2, x_data_d2, second_derivative)

plt.plot(x_fine_d2, second_deriv_fine, label="Друга похідна y''(x) (2. спосіб)", color='green')
plt.title("Друга похідна y''(x)")
plt.xlabel('x')
plt.ylabel("y''")
plt.legend()
plt.grid(True)

# Показуємо графіки
plt.tight_layout()
plt.show()
```
